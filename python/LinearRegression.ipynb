# Linear Regression

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import regression as reg
```

Make up some correlated data by adding noise to a quadratic curve.

```python
m = 100
x = np.random.random(m)
y = 2.4 + 3.2*x + 50.0*x**2 + 5*np.random.randn(m)
```

Define a function that will train the regression model and plot the data and curves.

```python
def train_and_plot(x, N, m_plt=100, tol=0.0001, verbose=True):
    lr = reg.LinearRegression()

    X = np.vander(x,N+1,increasing=True)
    x_plt = np.linspace(np.min(x),np.max(x),m_plt)

    theta_gd = lr.train(X,y,max_iters=1000,tol=0.0001,verbose=verbose)
    output_gd = lr.predict(X)
    y_plt_gd = lr.predict(np.vander(x_plt,N+1,increasing=True))

    theta_nm = lr.solve_normal_eqns(X,y)
    output_nm = lr.predict(X)
    y_plt_nm = lr.predict(np.vander(x_plt,N+1,increasing=True))

    plt.scatter(x,y,label='Data')
    plt.plot(x_plt,y_plt_gd,label='Grad Desc')
    plt.plot(x_plt,y_plt_nm,label='Normal')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    return lr
```

```python
lr = train_and_plot(x,1,2)
```

```python
plt.semilogy(lr.vtheta_err,'o',label='Theta error')
plt.semilogy(lr.vcost,'o',label='Cost')
plt.grid()
plt.xlabel('Iteration')
plt.legend()
```
