# Linear Regression

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import regression as reg
```

Make up some correlated data by adding noise to a quadratic curve.

```python
m = 100
x = np.random.random(m)
y = 2.4 + 3.2*x + 50.0*x**2 + 5*np.random.randn(m)

idx = np.argsort(x)
x = x[idx]
y = y[idx]
```

Fit a straight line.

```python
lr = reg.LinearRegression()

X = np.hstack((np.ones((m,1)),x.reshape((m,1))))

theta_gd = lr.train(X,y,max_iters=500,tol=0.0001,verbose=True)
output_gd = lr.predict(X)

theta_nm = lr.solve_normal_eqns(X,y)
output_nm = lr.predict(X)

plt.scatter(x,y)
plt.plot(x,output_gd)
plt.plot(x,output_nm)
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.semilogy(lr.vtheta_err,'o',label='Theta error')
plt.semilogy(lr.vcost,'o',label='Cost')
plt.grid()
plt.xlabel('Iteration')
plt.legend()
```

Fit a quadratic curve.

```python
lr = reg.LinearRegression()

X = np.hstack((np.ones((m,1)),x.reshape((m,1)),(x**2).reshape((m,1))))
theta = lr.train(X,y)
output = lr.predict(X)

plt.scatter(x,y)
plt.plot(x,output)
plt.xlabel('x')
plt.ylabel('y')
```
